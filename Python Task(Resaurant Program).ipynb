```python

def restaurant_menu():
    # Menu items with their prices
    starters = {
        'Chesse balls': 150.00,
        'Paneer Swasthik': 100.00,
        'Chesse Rings': 50.00
    }

    main_courses = {
        'Biryani':500.00,
        'Butter chicken': 400.00,
        'Paav Bhaji': 300.00
    }

    desserts = {
         'Ice Cream': 80.00,
        'Cake': 100.00,
        'Faluda': 50.00
    }

    total_bill = 0.00  # Initialize total bill
    while True:
        # Display the menu
        print("\nWelcome to the Restaurant!")
        print("\n--- Starters ---")
        for item, price in starters.items():
            print(f"{item}: ₹{price:.2f}")

        print("\n--- Main Courses ---")
        for item, price in main_courses.items():
            print(f"{item}: ₹{price:.2f}")

        print("\n--- Desserts ---")
        for item, price in desserts.items():
            print(f"{item}: ₹{price:.2f}")

        # User selects items
        starter_choice = input("\nEnter the starter you want: ")
        main_course_choice = input("Enter the main course you want: ")
        dessert_choice = input("Enter the dessert you want: ")

        # Validate the user's choices and get the prices
        if starter_choice in starters:
            starter_price = starters[starter_choice]
        else:
            print("Invalid starter choice! Defaulting to ₹0.00")
            starter_price = 0.00

        if main_course_choice in main_courses:
            main_course_price = main_courses[main_course_choice]
        else:
            print("Invalid main course choice! Defaulting to ₹0.00")
            main_course_price = 0.00

        if dessert_choice in desserts:
            dessert_price = desserts[dessert_choice]
        else:
            print("Invalid dessert choice! Defaulting to ₹0.00")
            dessert_price = 0.00

        # Calculate the total bill
        meal_total = starter_price + main_course_price + dessert_price
        total_bill += meal_total

        # Display the current bill for the selected items
        print("\n--- Selected Meal Summary ---")
        print(f"Starter ({starter_choice}): ₹{starter_price:.2f}")
        print(f"Main Course ({main_course_choice}): ₹{main_course_price:.2f}")
        print(f"Dessert ({dessert_choice}): ₹{dessert_price:.2f}")
        print(f"Meal Total: ₹{meal_total:.2f}")
        print("------------------------------")
        print(f"Running Total Bill: ₹{total_bill:.2f}")

        # Ask the user if they want to order more
        more_order = input("\nWould you like to order more? (yes/no): ").strip().lower()
        if more_order != 'yes':
            break

    # Final total bill
    print("\n--- Final Bill Summary ---")
    print(f"Total Bill: ₹{total_bill:.2f}")
    print("Thank you for dining with us!")

# Call the function
restaurant_menu()
```
